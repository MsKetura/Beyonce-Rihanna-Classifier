```python
import lyricsgenius as lg
```

```python
file = open('capstone_lyrics.txt', "w")
```

```python
genius = lg.Genius('DryVlUTbteHaKHVcEbfawg6NCeRUZxNPPB43VPEe5IdZ5usYzueffeRADPL2-EJu', 
                   skip_non_songs=True, 
                   excluded_terms=["(Remix)", "(Live)"], 
                   remove_section_headers=True)
```

```python
artists = ['Beyonce', 'Rihanna']
```

```python
def get_lyrics(arr, k):
    c = 0
    for name in arr:
        try:
            songs = (genius.search_artist(name, max_songs=75, sort='popularity')).songs
            s = [song.lyrics for song in songs]
            file.write("\n \n   <|endoftext|>   \n \n".join(s))
            c += 1
            print(f"Songs grabbed:{len(s)}")
        except:
            print(f"some exception at {name}: {c}")
            
get_lyrics(artists, 75)
```
